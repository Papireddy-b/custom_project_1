```python
teens=pd.read_csv('NCHS_-_Teen_Birth_Rates_for_Females_by_Age_Group__Race__and_Hispanic_Origin__United_States.csv')
pregnancy=pd.read_csv('NCHS_-_Pregnancy_Rates__by_Age_for_Hispanic_Women__United_States__1990-2010.csv')
Drugs=pd.read_csv('NCHS_-_Drug_Poisoning_Mortality_by_County__United_States.csv')
mortality=pd.read_csv('NCHS_-_Infant_Mortality_Rates__by_Race__United_States__1915-2013.csv')
```

```python
teens.head()
```

```python
teens.rename(columns={'Race or Hispanic Origin':'Race Ethnicity'},inplace=True)
teens.head()
```

```python
teens['Year']=pd.to_datetime(teens.Year)
pregnancy['Year']=pd.to_datetime(pregnancy.Year)
Drugs['Year']=pd.to_datetime(Drugs.Year)
mortality['Year']=pd.to_datetime(mortality.Year)
```

# Join teens to pregnancy, keeping only rows from transactions and users that match via Race Ethnicity(inner join) 

```python
pregnancy.merge(teens,how='left',on='Race Ethnicity')
```

# Which pregnancy have a Race Ethnicity not in teens? 

```python
pregnancy[~pregnancy['Year'].isin(teens['Year'])]
```

# Join teens to pregnancy, keeping only rows from pregnancy and teens that match via Race Ethnicity(inner join) 

```python
pregnancy.merge(teens,how='inner',on='Race Ethnicity')
```

# Join teens to pregnancy, displaying all matching rows AND all non-matching rows (full outer join)

```python
pregnancy.merge(teens,how='outer',on='Race Ethnicity')
```

# Determine  different  Age Groups marrital Status on same year

```python
pd.merge(left=teens, right=pregnancy, how='inner', left_on=['Year', 'Age Group'], right_on=['Year', 'Marrital Status'])
```
